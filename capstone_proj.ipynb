```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

#creating source dataframes
gini = pd.read_csv("API_SI.POV.GINI_DS2_en_csv_v2_53.csv", delimiter=',', dtype=str) #Gini Index by country
countries = pd.read_csv("Metadata_Country_API_SI.POV.GINI_DS2_en_csv_v2_53.csv", delimiter=',', dtype=str) #Country details, specifies high to low income population.
econ_crime = pd.ExcelFile("data_cts_corruption_and_economic_crime.xlsx")
violent_crime = pd.ExcelFile("data_cts_violent_and_sexual_crime.xlsx")
homicide_crime = pd.ExcelFile("data_cts_intentional_homicide.xlsx")

econ = econ_crime.parse()  
violent = violent_crime.parse()
homicide = homicide_crime.parse()

#removing top two rows for all Excel files, setting Row 1 as column headers 
econ = pd.DataFrame(econ.values[2:], columns=econ.iloc[1])
violent = pd.DataFrame(violent.values[2:], columns=violent.iloc[1])
homicide = pd.DataFrame(homicide.values[2:], columns=homicide.iloc[1])

#removing NaN values
gini = gini.fillna("Unknown")
countries = countries.fillna("Unknown")
econ = econ.fillna("Unknown")
violent = violent.fillna("Unknown")
homicide = homicide.fillna("Unknown")

display(gini.head(5))
display(countries.head(5))
display(econ.head(5))
display(violent.head(5))
display(homicide.head(5))
```

NOTES: 

Gini .csv files pulled from https://data.worldbank.org/indicator/SI.POV.GINI (World Bank)

Crime data pulled from https://dataunodc.un.org/ (United Nations Office on Drugs and Crime)
